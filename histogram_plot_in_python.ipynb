```python
import seaborn as sns
penguins = sns.load_dataset("penguins")
print(penguins)
```

```python
sns.histplot(data=penguins, x="flipper_length_mm")
```

```python
sns.histplot(data=penguins, x="flipper_length_mm", binwidth=3)
```

```python
# hue: Semantic variable that is mapped to determine the color of plot elements
sns.histplot(data=penguins,hue='species', x="flipper_length_mm", binwidth=3)
```

```python
# kde: If True, compute a kernel density estimate
sns.histplot(data=penguins,hue='species', x="flipper_length_mm", binwidth=3,kde=True)
```
